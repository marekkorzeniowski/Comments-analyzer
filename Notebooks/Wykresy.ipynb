```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')
%matplotlib inline

# set jupyter's max row display
pd.set_option('display.max_row', 1000)

# set jupyter's max column width to 50
pd.set_option('display.max_columns', 50)
```

```python
df1 = pd.read_csv('pos_01.csv', index_col='sentiment_label').sort_values(by=['Counter'])
df1
```

```python
## 1

ax = df1.plot(kind='barh', figsize=(10,7),color="coral", fontsize=13, legend=None);
ax.set_alpha(0.8)
ax.set_title("Rozkład sentymentu dla postów", fontsize=18)
ax.set_xlabel("Ilość", fontsize=18);
ax.set_xticks([0, 7500, 15000, 22500, 30000, 37500])
ax.set_ylabel(None)

# create a list to collect the plt.patches data
totals = []

# find the values and append to list
for i in ax.patches:
    totals.append(i.get_width())

# set individual bar lables using above list
total = sum(totals)

# set individual bar lables using above list
for i in ax.patches:
    # get_width pulls left or right; get_y pushes up or down
    ax.text(i.get_width()+.3, i.get_y()+.38, \
            str(round((i.get_width()/total)*100, 2))+'%', fontsize=15,
color='dimgrey')

# invert for largest on top 
#ax.invert_yaxis()
plt.savefig("wykres_1_posts.png")
```
